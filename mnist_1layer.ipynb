```python
import scipy.special
import numpy as np
import matplotlib.pyplot
from sklearn.datasets import fetch_openml
%matplotlib inline
```

```python
mnist = fetch_openml('mnist_784')
```

```python
mnist.data.shape, mnist.target.shape
```

```python
data = mnist.data
target = mnist.target
```

```python
np.shape(data)
```

```python
target1 = target.astype(str).astype(int)
```

```python
target1
```

```python
def onehot(labels,classes):
    y = np.eye(classes)
    return y[labels]


target2 = onehot(target1,10)
```

```python
target2
```

```python
x = data[:30000] 
```

```python
y = target2[:30000]
```

```python
XT = x.T
```

```python
R1 = np.random.randn(30000,500)
```

```python
W1 = XT@R1
np.shape(W1)
```

```python
O1 = x@W1
O1
```

```python
FO1 = np.arctan(O1)
FO1
```

```python
aa = np.linalg.pinv(FO1)
```

```python
w2 = aa@y
```

```python


aaa = np.linalg.pinv(w2)

np.shape(aaa)
```

```python
NEWW1 = np.linalg.pinv(x)@(np.tan(y@aaa))
```

```python
NEWW2 = (np.linalg.pinv(np.arctan(x@NEWW1)))@y
```

```python
plz = (np.arctan(x@NEWW1))@NEWW2
```

```python
np.argmax(plz,axis =1)
```

```python
target1[:30000]
```

```python
xtest = data[35000:40000] 
```

```python
len(xtest)
```

```python
a = np.arctan(xtest@NEWW1)
```

```python
plz2 = a@NEWW2
```

```python
plz2[0]
```

```python
predict = np.argmax(plz2,axis =1)
predict = predict.tolist()
predict
```

```python
testy = target1[35000:40000]
testy = testy.tolist()
```

```python
testy
```

```python
scorecard = []

for i in range(len(xtest)):
    correct_label = testy[i]
    label = predict[i]
    print(label)
    if(label == correct_label):
        scorecard.append(1)
    else:
        scorecard.append(0)
        pass
    pass
```

```python
scorecard_array = np.asarray(scorecard)
print("performance = ", scorecard_array.sum() / scorecard_array.size)
```
